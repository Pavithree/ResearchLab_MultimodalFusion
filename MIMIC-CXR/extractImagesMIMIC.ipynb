```python
folders = glob.glob('DataSet/MIMIC/physionet.org/files/mimic-cxr-jpg/2.0.0/files/p19/p*/s*')
imagenames_list = []
for folder in folders:
    for f in glob.glob(folder+'/*.jpg'):
        imagenames_list.append(f)

# print(imagenames_list[0])
image_size = 256

patientSourceId = []
radiologyChestXrayImages = []        
for image in imagenames_list:
    patientSourceId.append(image.split("/")[9])
    image_read = cv2.imread(image)
    image_resized = cv2.resize(image_read,(image_size,image_size))
    radiologyChestXrayImages.append(image_resized)
    cv2.destroyAllWindows()
    print(image)
    
info = {"images":radiologyChestXrayImages ,
       "patientId": patientSourceId}

df = pd.DataFrame(info)

df.to_pickle('DataSet/CSV/radiologyImagesP19.pkl')
```

```python
completeDf = pd.DataFrame()
for i in range(10):
    with open('DataSet/CSV/radiologyImagesP1'+str(i)+'.pkl', 'rb') as f:
        data = pickle.load(f)
#         print(data.shape)
        completeDf = completeDf.append(data, ignore_index=True)
        print(completeDf.shape)
```
