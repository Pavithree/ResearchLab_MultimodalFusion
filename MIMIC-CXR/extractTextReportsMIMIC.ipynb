```python
import glob
import re
import pandas as pd
```

```python
# files = glob.glob ("multimodal\\Data\\radiologyReports\\p*\\p*\\*.txt")
files = glob.glob ("path to text reports")


patientIdList = []
patientRadiologyReport = []

for file in files:
    data = ' '
    patientId = file.split("\\")[5].split(".")[0]
    patientIdList.append(patientId)
    fileName = file.split("\\")[3]
    with open(file, 'r') as file:
        data = file.read().replace('\n', '') 
        
        #if finding is present 
        if data.find("FINDINGS:") != -1:
            start = data.find("FINDINGS:") + len("FINDINGS:")
            end = data.find("IMPRESSION:")
            findings = data[start:end]
        else:
            findings = " "
            
        if data.find("IMPRESSION:") != -1:
            impression = data.split("IMPRESSION:")[1]
        else:
            impression = " "
            
        substring = findings + impression
       
        
        patientRadiologyReport.append(substring.strip())
```

```python
patientDetails = list(zip(patientIdList,patientRadiologyReport))
df = pd.DataFrame(patientDetails, columns=['SourceId','RadiologyReport'])
df
```

```python
files = glob.glob ("MultimodalMLHealthcareDomain\\Data\\radiologyReports\\p*\\p*\\*.txt")

patientIdListFindingsImpression = []
patientRadiologyReportFindingsImpression = []

for file in files:
    data = ' '
    patientId = file.split("\\")[5].split(".")[0]
    patientIdListFindingsImpression.append(patientId)
    fileName = file.split("\\")[5]
    with open(file, 'r') as file:
        data = file.read().replace('\n', '') 
        if data.find("FINDINGS:") != -1:
            start = data.find("FINDINGS:") + len("FINDINGS:")
            end = data.find("IMPRESSION:")
            findings = data[start:end]
        else:
            findings = " "
            
        if data.find("IMPRESSION:") != -1:
            impression = data.split("IMPRESSION:")[1]
        else:
            impression = " "
            
        substring = findings + impression
        
        patientRadiologyReportFindingsImpression.append(substring.strip())
```
